# 3. Upscale video

- The upscaled video will be saved to `output_dir`
- If google drive is mounted, it will be also saved at `MyDrive/Upscaled Videos (REAL-ESRGAN)`


```python
import os
import cv2

# --- Parameters ---
video_name = "Young_professional_working_202603291300.mp4"
video_path = f"/content/{video_name}"
output_dir = "/content/"
model = "RealESRGAN_x4plus"

if not os.path.exists(video_path):
    print(f"❌ Error: {video_name} ෆයිල් එක නැහැ! කරුණාකර Upload කරන්න.")
else:
    # Scale calculation (4K target)
    video_capture = cv2.VideoCapture(video_path)
    w = int(video_capture.get(cv2.CAP_PROP_FRAME_WIDTH))
    h = int(video_capture.get(cv2.CAP_PROP_FRAME_HEIGHT))
    scale_factor = max(3840/w, 2160/h)

    print(f"🚀 Upscaling to 4K... ({w}x{h} -> 3840x2160)")

    # Real-ESRGAN Run කරනවා
    # --half අයින් කළා, --tile 512 තියෙන්නේ memory බේරගන්න
    !python inference_realesrgan_video.py -n {model} -i '{video_path}' -o '{output_dir}' --outscale {scale_factor} -t 512

    # Rename file
    base_name = os.path.splitext(video_name)[0]
    old_file = os.path.join(output_dir, f"{base_name}_out.mp4")
    final_file = os.path.join(output_dir, f"{base_name}_4K_READY.mp4")

    if os.path.exists(old_file):
        os.rename(old_file, final_file)
        print("-" * 30)
        print(f"🎉 WADE HARI MACHAN! Download here: {final_file}")
        print("-" * 30)
    else:
        print("❌ Upscale එකේ අවුලක් තියෙනවා. Logs බලන්න.")
```
